```python
import random
population = []

for i in range(1000):
  p = random.randint(0,100)
  population.append(p)

print(population)
```

```python
import matplotlib.pyplot as plt
plt.hist(population,bins=30,color='cyan',edgecolor='black')
plt.title('Histogram of Population')
plt.xlabel('values')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
mean_population = np.mean(population)
variance_population = np.var(population)
print("Mean of Population is",mean_population)
print("Variance of population is",variance_population)
```

```python
sample_size = 50
sample = random.sample(population,sample_size)
print("Random Sample",sample)
```

```python
import numpy as np
mean_value = 50
std_deviation = 10
population = np.random.normal(loc=mean_value,scale=std_deviation,size=1000)
print("Generated Population(first 10 elements)",population[:10])
```

```python
import numpy as np
mean_population = np.mean(population)
mean_population = np.mean(population)
variance_population = np.var(population)
print("Mean of Population is",mean_population)
print("Variance of population is",variance_population)
```

```python
import matplotlib.pyplot as plt
import numpy as np
population = np.random.rand(100)
num_samples = 1000
sample_size = 30
sample_means = np.zeros(num_samples)
for i in range(num_samples):
  sample = np.random.choice(population,size=sample_size)
  sample_means[i]=np.mean(sample)

plt.hist(sample_means,bins=30,edgecolor='black')
plt.title('Sampling distribution of Sample Mean (Non Normal population)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
population = np.random.normal(loc=0,scale=1,size=1000)
num_samples = 1000
sample_size = 30
sample_means = np.zeros(num_samples)
for i in range(num_samples):
  sample=np.random.choice(population,size=sample_size)
  sample_means[i] = np.mean(sample)

plt.hist(sample_means,bins=30,color='green',edgecolor='black')
plt.title('Sampling distribution of Sample Mean (Normal population)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
true_propotion = 0.395
population_size = 1000
num_samples = 1000
sample_size = 30
sample_propotion = np.zeros(num_samples)
for i in range(num_samples):
  sample=np.random.choice([0,1],p=[1-true_propotion,true_propotion],size=sample_size)
  sample_propotion[i] = np.mean(sample)

plt.hist(sample_propotion,bins=30,color='magenta',edgecolor='black')
plt.title('Sampling distribution of Sample Mean (Normal population)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
